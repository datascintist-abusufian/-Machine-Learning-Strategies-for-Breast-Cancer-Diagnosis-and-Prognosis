```python
import matplotlib.pyplot as plt

# Class names and their corresponding probabilities
classes = ['benign', 'malignant', 'normal']
probs = [0.020313283428549767, 0.9778401255607605, 0.001846558996476233]

# Create a bar chart with new colors
colors = ['lightblue', 'lightcoral', 'lightgreen']
plt.bar(classes, probs, color=colors)

# Set labels, title, and their font properties
plt.xlabel('Classes', fontweight='bold')
plt.ylabel('Probability', fontweight='bold')
plt.title('Prediction Probabilities for the Image', fontweight='bold')
plt.ylim(0, 1.1)  # set y-axis limits to [0, 1.1] for clarity

# Display the exact probabilities on top of the bars
for i, v in enumerate(probs):
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center', va='bottom', fontsize=10, fontweight='bold')

# Highlight the class with the highest probability
predicted_class = "malignant"
plt.gca().get_children()[classes.index(predicted_class)].set_color('gold')

# Enhance the visibility of the axis' tick labels
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')

plt.show()
```
